```python
import pandas as pd
```

```python
# Display all columns when printing DataFrames (prevents truncation)

pd.set_option('display.max_columns', None)
```

```python
# Read the Star Wars survey CSV file
# cp1252 encoding is used to handle special characters from Windows/Excel

df = pd.read_csv('../Data/raw/StarWars.csv', encoding= 'cp1252')
```

```python
# Preview the first few rows of the raw dataset

df.head()
```

```python
# Display all column names in the dataset

df.columns
```

```python
# Rename columns to meaningful, readable names
# This fixes "Unnamed" columns caused by multi-row headers in the CSV
# and cleans an encoding issue in one column name

df_renamed = df.rename(columns={
    'Which of the following Star Wars films have you seen? Please select all that apply.' : 'Have you seen Star Wars: Episode I The Phantom Menace?',
    'Unnamed: 4' : 'Have you seen Star Wars: Episode II Attack of the Clones?', 
    'Unnamed: 5' : 'Have you seen Star Wars: Episode III Revenge of the Sith?', 
    'Unnamed: 6' : 'Have you seen Star Wars: Episode IV A New Hope?', 
    'Unnamed: 7' : 'Have you seen Star Wars: Episode V The Empire Strikes Back?', 
    'Unnamed: 8' : 'Have you seen Star Wars: Episode VI Return of the Jedi?',
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.' : 'Ranking of Episode 1',
    'Unnamed: 10': 'Ranking of Episode 2',  
    'Unnamed: 11' : 'Ranking of Episode 3', 
    'Unnamed: 12' : 'Ranking of Episode 4', 
    'Unnamed: 13' : 'Ranking of Episode 5',
    'Unnamed: 14' : 'Ranking of Episode 6',
    'Please state whether you view the following characters favorably, unfavorably, or are unfamiliar with him/her.': 'Are you familiar with Han Solo?',
    'Unnamed: 16': 'Are you familiar with Luke Skywalker?', 
    'Unnamed: 17': 'Are you familiar with Princess Leia Organa?', 
    'Unnamed: 18': 'Are you familiar with Anakin Skywalker?', 
    'Unnamed: 19': 'Are you familiar with Obi Wan Kenobi?',
    'Unnamed: 20': 'Are you familiar with Emperor Palpatine?', 
    'Unnamed: 21': 'Are you familiar with Darth Vader?', 
    'Unnamed: 22': 'Are you familiar with Lando Calrissian?', 
    'Unnamed: 23': 'Are you familiar with Boba Fett?',
    'Unnamed: 24': 'Are you familiar with C-3P0?', 
    'Unnamed: 25': 'Are you familiar with R2 D2?', 
    'Unnamed: 26': 'Are you familiar with Jar Jar Binks?', 
    'Unnamed: 27': 'Are you familiar with Padme Amidala?',
    'Unnamed: 28': 'Are you familiar with Yoda?',
    'Do you consider yourself to be a fan of the Expanded Universe?Œæ' : 'Do you consider yourself to be a fan of the Expanded Universe?'
})
```

```python
# Define critical columns that must not contain missing values

important_cols = [
    'RespondentID',
    'Have you seen any of the 6 films in the Star Wars franchise?',
    'Gender',
    'Age'
]

# Drop rows where any of the important columns contain NaN values

df_renamed.dropna(subset=important_cols, inplace=True)
```

```python
# Preview cleaned DataFrame

df_renamed.head()
```

```python
# Check updated column names after cleaning

df_renamed.columns
```

```python
# List of movie "seen" columns to convert into Yes/No values

to_repalce_values =['Have you seen Star Wars: Episode I The Phantom Menace?',
       'Have you seen Star Wars: Episode II Attack of the Clones?',
       'Have you seen Star Wars: Episode III Revenge of the Sith?',
       'Have you seen Star Wars: Episode IV A New Hope?',
       'Have you seen Star Wars: Episode V The Empire Strikes Back?',
       'Have you seen Star Wars: Episode VI Return of the Jedi?' 
]

# Convert movie columns:
# If a value exists → "Yes"
# If NaN → "No"
# Convert into datatype category

for col in to_repalce_values:
    df_renamed[col] = df_renamed[col].notna().map({True: 'Yes', False: 'No'}).astype('category')
```

```python
# Columns containing movie rankings

ranking_cols = [
    'Ranking of Episode 1', 'Ranking of Episode 2',
    'Ranking of Episode 3', 'Ranking of Episode 4',
    'Ranking of Episode 5', 'Ranking of Episode 6'
]

# Convert ranking columns to numeric values
# Invalid values are coerced into NaN

df_renamed[ranking_cols] = df_renamed[ranking_cols].apply(pd.to_numeric, errors= 'coerce')

# Extract and sort unique ranking values across all ranking columns

unique_values = pd.unique(df_renamed[ranking_cols].values.ravel())
sorted(unique_values)
```

```python
# View distribution of Age values

df_renamed['Age'].value_counts()
```

```python
# View distribution of Household Income values

df_renamed['Household Income'].value_counts()
```

```python
# Categorize Age and Household Income into readable groups
# Convert category cols into datatype category

category_cols = [['Age', 'Household Income']]

for col in category_cols:

    df_renamed[col] =  df_renamed[col].replace({
            '18-29' : 'Young Adult',
            '30-44' : 'Adult',
            '45-60' : 'Middle-Aged',
            '> 60'  : 'Senior',
            '$0 - $24,999' : 'Low Income',
            '$25,000 - $49,999': 'Lower-Middle Income',
            '$50,000 - $99,999' : 'Middle Income',
            '$100,000 - $149,999' : 'Upper-Middle Income',
            '$150,000+' : 'High Income'
    }).astype('category')
```

```python
# Check categorized Age values

df_renamed['Age'].value_counts()
```

```python
# Check categorized Household Income

df_renamed['Household Income'].value_counts()
```

```python
# Drop an irrelevant or unused column

df_renamed = df_renamed.drop('Which character shot first?', axis=1)
```

```python
# Save the cleaned dataset to a CSV file
# Index=False prevents pandas from writing the index as a column

df_renamed.to_csv('../Data/processed/star_wars_survey_cleaned.csv', index= False)
```
